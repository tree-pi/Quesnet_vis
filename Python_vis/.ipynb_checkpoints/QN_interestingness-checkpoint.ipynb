```python
# part 1 
# basic definition 

import networkx as nx 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from pyvis.network import Network
import json 

from PyQt5.QtCore import QDateTime, Qt, QTimer
from PyQt5.QtWidgets import (QApplication, QCheckBox, QComboBox, QDateTimeEdit,
        QDial, QDialog, QGridLayout, QGroupBox, QHBoxLayout, QLabel, QLineEdit,
        QProgressBar, QPushButton, QRadioButton, QScrollBar, QSizePolicy,
        QSlider, QSpinBox, QStyleFactory, QTableWidget, QTabWidget, QTextEdit,
        QVBoxLayout, QWidget, QToolTip, QTableWidget, QTableWidgetItem)
from PyQt5.QtGui import QIcon, QPixmap
from PyQt5.QtGui import QFont
```

```python
# part 2
# import the data 

# with open('QNnodecuriosity.txt', 'rb') as handle:
#     b = pickle.loads(handle.read())
#     print(b)

allnode = []
# QNnodes.append(line)
with open('QNnodeinterestingness1.txt', 'r') as f:
    for line in f:
        data = json.loads(line)
        allnode.append(data)
# print(allnode)

alllink = []
# QNnodes.append(line)
with open('QNlinkinterestingness1.txt', 'r') as f:
    for line in f:
        data = json.loads(line)
#         print(data)
        alllink.append(data)
# print(alllink)
#         data.append(json.loads(line))
#     datastore = json.load(f)
```

```python
# part 3 
# preview the graph 

Q = nx.Graph()

# add nodes into the graph
for node in allnode:
#     print(node)
    nd_index = node['id']
    nd_label = node['feature']
#     nd_label = 'la'
    Q.add_node(nd_index,label=nd_label)

# add edges into the graph
for link in alllink:
    link_from = link['start']
    link_to = link['end']
#     nd_index = node['id']
    link_label = link['feature']
#     nd_label = 'la'
    Q.add_edge(link_from,link_to,label=link_label)
    
fig = plt.figure(frameon=False)
fig.set_size_inches(28, 12) 
ax = fig.add_axes([0, 0, 1, 1])
ax.axis('off')

pos = nx.kamada_kawai_layout(Q)
# nx.draw(Q,pos)

# set the color map 
# color_map = []
# for node in allnodes:
#     if node['id'] == rid:
#         color_map.append('orange')
#     elif node['id'][0] == 'a':
#         color_map.append('green')
#     else:
#         color_map.append('blue')

nx.draw_networkx(Q,pos,arrows=True,arrowstyle='->',arrowsize=25,with_labels=False)
node_labels = nx.get_node_attributes(Q,'label')
# nx.draw_networkx_labels(Q,pos,labels=node_labels,font_size = 24)
edge_labels = nx.get_edge_attributes(Q,'label')
# nx.draw_networkx_edge_labels(Q, pos, labels = edge_labels,font_size = 12)
# nx.draw(Q,with_labels = True,font_weight = 'bold',node_size = 1000,node_color='green',node_shape='h')

# save and illustration
# fig.savefig('QuestionNet8.png', dpi=100)
plt.show()
```

```python
# part 4
# generate the html version of the graph

g = Network(height="750px", width="100%", bgcolor="#222222", font_color="white")
# g.add_node(0)
# g.add_node(1)
# g.add_edge(0, 1)

g.from_nx(Q)
g.toggle_drag_nodes(True)
# g.toggle_stabilization(True)
g.toggle_hide_edges_on_drag(True)
g.show("interesting1.html")
```

```python
# condense paper html generation 

# The program aims to generate the html file from the 
# condensed topic graph

import networkx as nx 
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import json 
import pickle
import os

from IPython.core.debugger import set_trace
```

```python
# illustrate the topic network 

# extract from the paper network 

# pickfile=open(fdir+'%s'%file_list[i],'rb')
fdir = 'database/curiosity/algorithms/'
pickfile=open(fdir+'combQnet1.p','rb')
mydict = pickle.load(pickfile)
pickfile.close()

# divde the node and link 
[dic_node, dic_link] = mydict

# graph initialization 
topic_graph = nx.Graph()
# html json initialization
ht_nodes = []
ht_edges = []


# add nodes into the graph
(row, col) = dic_node.shape

for index in range(row):
    nd_id = dic_node.iloc[index]['id']
    nd_label = dic_node.iloc[index]['label']
    nd_paper = dic_node.iloc[index]['paper']
    
    num_paper = len(nd_paper)

    nd_size = 10*num_paper 
    if num_paper == 1:
        nd_color = 'blue'        
    elif num_paper == 2:
        nd_color = 'green' 
    else:
        nd_color = 'tomato'
        
    ht_nodes.append({'id':nd_id, 'label': nd_label, 'paper':nd_paper,
                     'color': nd_color, 'size':nd_size})

# add link into the graph 
(row, col) = dic_link.shape

for index in range(row):
    lk_id = dic_link.iloc[index]['id']
    lk_from = dic_link.iloc[index]['startNend'][0]
    lk_to = dic_link.iloc[index]['startNend'][1]
    lk_label = dic_link.iloc[index]['label'][0]
    lk_paper = dic_link.iloc[index]['paper']
    
    ht_edges.append({'from':lk_from, 'to': lk_to, 'label': lk_label,
                     'font':{'align':'middle'},
                     'arrows':'to, middle'})

# export the json file for the html file 
data = []
data = {"nodes":ht_nodes,"edges":ht_edges}

with open('jscode/format/jscombine.json', 'w') as outfile:  
    json.dump(data, outfile)
```
